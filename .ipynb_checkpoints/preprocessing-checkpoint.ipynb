```python
# Δεδομένου ότι η πληροφορία για το έτος κατασκευής του σπιτιού περιλαμβάνεται ήδη (YearBuilt),
# γίνεται ομαδοποίηση των τιμών του MSSubClass βάσει του αριθμού των επιπέδων (stories),
# με στόχο τη μείωση της διαστατικότητας (dimensionality).

def simplify_mssubclass(val):
    if val in [20, 30, 40, 120]:       # 1-story
        return '1-Story'
    elif val in [45, 50, 150]:         # 1.5-story
        return '1.5-Story'
    elif val in [60, 70, 160]:         # 2-story
        return '2-Story'
    elif val in [75]:                  # 2.5-story
        return '2.5-Story'
    elif val in [80, 85, 180]:         # Split-level or foyer
        return 'Split/Multi-Level'
    elif val in [90, 190]:             # Duplex/Conversion
        return 'Multi-Family'
```

```python
df['MSSubClass'] = df['MSSubClass'].apply(simplify_mssubclass)
```

```python
# Θεωρώντας ότι υπάρχει εν δυνάμει γραμμική σχέση μεταξύ του τύπου της κατοικίας και της τιμής πώλησης,
# εφαρμόζεται αριθμητικό mapping στις κατηγορίες, ώστε να είναι αξιοποιήσιμες από αλγορίθμους regression.

story_mapping = {
    '1-Story': 0,
    '1.5-Story': 1,
    '2-Story': 2,
    '2.5-Story': 3,
    'Split/Multi-Level': 4,
    'Multi-Family': 5
}
```

```python
df['MSSubClass'] = df['MSSubClass'].map(story_mapping)
```

```python
# Η εποχή θα κωδικοποιηθεί μέσω ομαδοποίησης (clustering) σε n_clusters ομάδες (υπερπαράμετρος),
# με σκοπό τη μείωση της πολυπλοκότητας του χαρακτηριστικού.
# Δεδομένου ότι θα χρησιμοποιηθεί Linear Regression,
# το χαρακτηριστικό αυτό θα μετατραπεί σε μορφή one-hot encoding.

plt.scatter(df.MoSold, df.SalePrice)
```

```python
def cluster_weeks(df, n_clusters=4):
    '''
    Κάνει mapping τους μήνες σε γωνίες και τις γωνίες σε νεο 2D χώρο (cos, sin)
    Κανονικοποιεί το target και κάνει KMeans clustering στον 3D χωρο (cos, sin, target)
    '''
    df_copy = df.copy()
    df_copy['SalePriceNorm'] = (df_copy['SalePrice']-df_copy['SalePrice'].mean())/df_copy['SalePrice'].std()
    
    data = df_copy.groupby('MoSold')['SalePriceNorm'].mean().reset_index()
    
    n_clusters = n_clusters
    
    months = data['MoSold'].values
    theta = 2 * np.pi * months / 12 # Map weeks to angles
    x_sin = np.sin(theta)           # Sine transformation
    x_cos = np.cos(theta)           # Cosine transformation
    
    data['cos'] = x_cos
    data['sin'] = x_sin
    
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(data[['cos', 'sin', 'SalePriceNorm']])
    data['cluster'] = kmeans.predict(data[['cos', 'sin', 'SalePriceNorm']])
    
    # Mapping για κάθε μήνα
    # cos και sin μονο για visualization, μετά drop
    cluster_map = data.set_index('MoSold')['cluster'].to_dict()
    cos_map = data.set_index('MoSold')['cos'].to_dict()
    sin_map = data.set_index('MoSold')['sin'].to_dict()
    
    # Εφαρμογή στα δεδομένα
    df_copy['month_cluster'] = df_copy['MoSold'].map(cluster_map)
    df_copy['month_cos'] = df_copy['MoSold'].map(cos_map)
    df_copy['month_sin'] = df_copy['MoSold'].map(sin_map)
    
    df_copy.drop('SalePriceNorm', axis=1, inplace=True)

    return df_copy, data
```

```python
df, data = cluster_weeks(df)
```

```python
# Visualization των clusters χρησιμοποι΄ώντας το mean κάθε μήνα
plt.figure(figsize=(10, 5))
scatter = plt.scatter(data.MoSold, data.SalePriceNorm, c=data.cluster)

handles = []
for cluster_id in sorted(data['cluster'].unique()):
    color = scatter.cmap(scatter.norm(cluster_id))
    patch = mpatches.Patch(color=color, label=f'Cluster {cluster_id}')
    handles.append(patch)

plt.legend(handles=handles, title='Clusters')
plt.xlabel('Month Sold')
plt.ylabel('Standardized Sale Price')
plt.title('Clustered Months by Sale Price')
plt.savefig(f"{plot_dir}/monthslinear.png", bbox_inches ='tight')
plt.show()
```
